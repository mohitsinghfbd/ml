```python
from sklearn.neighbors import NearestNeighbors
```

```python
len(moviemat)
```

```python
moviemat1 = moviemat.fillna(0)
moviemat1.head()
```

```python
model_knn = NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(moviemat1)
```

```python
movie_list = movie_titles
movie_list.drop(movie_list.columns[[2,3,4]],axis=1,inplace=True)
movie_list
```

```python
def print_similar_movies(query_index) :
    query_index_movie_ratings = moviemat1.loc[query_index,:].values.reshape(1,-1)
    
    distances,indices = model_knn.kneighbors(query_index_movie_ratings,n_neighbors = 11) 
    for i in range(0,len(distances.flatten())):
        get_movie = movie_list.loc[movie_list['item_id']==query_index]['movie_title']
        if i==0:
            print('Recommendations for {0}:\n'.format(get_movie))
        else :
            indices_flat = indices.flatten()[i]
            get_movie = movie_list.loc[movie_list['item_id']==moviemat1.iloc[indices_flat,:].name]['movie_title']
            print('{0}: {1}, with distance of {2}:'.format(i,get_movie,distances.flatten()[i]))
```

```python
print_similar_movies(50)
```

```python
movie_content_df_temp = movie_titles.copy()
movie_content_df_temp.set_index('item_id')
movie_content_df = movie_content_df_temp.drop(columns = ['item_id','movie_title'])
movie_content_df = movie_content_df.as_matrix()
movie_content_df
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(movie_content_df,movie_content_df)
```

```python
cosine_sim
```

```python
indicies = pd.Series(movie_content_df_temp.index, movie_content_df_temp['movie_title'])
indicies
```

```python
def get_similar_movies_based_on_content(movie_index) :
    sim_scores = list(enumerate(cosine_sim[movie_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[0:11]
    print(sim_scores)
    movie_indices = [i[0] for i in sim_scores]
    print(movie_indices)
    similar_movies = pd.DataFrame(movie_content_df_temp[['movie_title']].iloc[movie_indices])
    return similar_movies
```
