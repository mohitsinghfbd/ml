```python
def cost(x,y,m,points):
    M=len(points)
    total_cost=0
    s=0
    for i in range(M):
            s=np.dot(m,x[i])
            #s=(m*x[i]).sum() -- equivalent to dot
            total_cost +=(1/M)*((y[i]-s )**2)
            #s=0                    
    return total_cost
        
```

```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(14)
    new_m=np.zeros(14)
    M=len(points)
    for i in range(M):
                  m_slope += np.dot(x[i], (y[i]-np.dot(m,x[i])))*(-2/M)
    # dont get confused ! in an np array(above) a+b is element wise addition     
    for j in range(14):    
         new_m[j]=m[j]-learning_rate*m_slope[j]
    return new_m
        
```

```python
def gd(points,learning_rate,num_iteration):
    m=np.zeros(14)
    for i in range(num_iteration):
        m=step_gradient(points,learning_rate,m)
        print(i ,'cost:',cost(x,y,m,points) )
        
    return m    
```
