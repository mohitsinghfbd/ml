```python
b=np.ones((len(data)),int).reshape(-1,1)
x=np.append(x,b,axis=1)
print(x.shape)
print(data_test.shape)
```

```python
def cost(x,y,m,points):
    M=len(points)
    total_cost=0
    s=0
    for i in range(M):
        s=np.dot(m,x[i])
        total_cost+=(1/M)*((y[i]-s)**2)
    return total_cost
```

```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(17)
    new_m=np.zeros(17)
    M=len(points)
    for i in range(M):
        m_slope+=np.dot(x[i],(y[i]-np.dot(m,x[i])))*(-2/M)
    for j in range(17):
        new_m[j]=m[j]-learning_rate*m_slope[j]
    return new_m
```

```python
def gd(points,learning_rate,num_iterations):
    m=np.zeros(17)
    
    for i in range(num_iterations):
        m=step_gradient(points,learning_rate,m)
        print(i,'cost:',cost(x,y,m,points))
        last=cost(x,y,m,points)
    return m
```

```python
def predict(m,data_test,y_pred):
    for i in range(len(data_test)):
        y_pred[i]=np.dot(m,data_test[i])
    return y_pred
```

```python
def run():
    num_iterations=1000
    learning_rate=0.01
    y_pred=np.zeros(len(data_test))
    m=gd(data,learning_rate,num_iterations)
    y_pred=predict(m,data_test,y_pred)
    np.savetxt('ans.csv',y_pred,delimiter=",",fmt="%.5f")
    print(m)
    print(y_pred.shape)
    print("HERE WE GO",y_pred)
```
