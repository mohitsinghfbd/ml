```python
a=np.ones((len(x)),int).reshape(-1,1)
x=np.append(x,a,axis=1)
x.shape
```

```python
def cost(x,y,m,points):
    M=len(points)
    total_cost=0
    s=0
    for i in range(M):
            s=np.dot(m,x[i])
            total_cost +=(1/M)*((y[i]-s )**2)                   
    return total_cost
```

```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(7)
    new_m=np.zeros(7)
    M=len(points)
    for i in range(M):
                  m_slope += np.dot(x[i], (y[i]-np.dot(m,x[i])))*(-2/M)    
    for j in range(7):    
         new_m[j]=m[j]-learning_rate*m_slope[j]
    return new_m
```

```python
def gd_desc(points,learning_rate,num_iteration):
    m=np.zeros(7)
    for i in range(num_iteration):
        m=step_gradient(points,learning_rate,m)
        print(i,' cost:',cost(x,y,m,points) )
        
    return m  
```

```python
def run():
    num_iteration=3000
    learning_rate=0.0000000000000000001
    y_pred=np.zeros(len(data_test))
    m=gd_desc(x,learning_rate,num_iteration)
    y_pred=prediction(m,data_test,y_pred)
    print(m)
    print(y_pred)
    return y_pred
```

```python
def prediction(m,data_test,y_pred):

    for i in range(len(data_test)):
            y_pred[i]=np.dot(m,data_test[i])
    return y_pred 
```
